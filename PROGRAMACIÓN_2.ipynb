En el intérprete, la salida de cadenas está encerrada en comillas y los caracteres especiales son escapados con barras invertidas. Aunque esto a veces luzca diferente de la entrada (las comillas que encierran pueden cambiar), las dos cadenas son equivalentes. Si no se quiere que los caracteres antepuestos por \ sean interpretados como caracteres especiales, se puede usar una cadena cruda agregando una r antes de la primera comilla.

```python
print('C:\\algun\\nombre')
print(r'C:\algun\nombre')
print(r'C:\algun\nombre')
```

Las cadenas de texto literales pueden contener múltiples líneas. Una forma es usar triple comillas: """ ... """ o ''' ... '''. Los fin de línea son incluidos automáticamente, pero es posible prevenir esto agregando una \ al final de la línea. Por ejemplo:

```python
print("""Uso:palabra 1
palabra 2
palabra 3""")
```

```python
print("""\
Uso:palabra 1
palabra 2
palabra 3
""")
```

En Python existen distintos elementos para guardar y organizar información, y cada uno se usa según la necesidad.

Variables : Se usan cuando queremos guardar un solo valor (un número, un texto, un resultadodo). Sirve para reutilizar datos y hacer el código más claro.

Tuplas: Se usan cuando queremos guarda varios valores, pero que no deban modificarse.Aportan seguridad pues los datos no cambian accidentalmente

Listas: Se usan cuando necesitamos guardar varios valores ordenados y que puedan cambiar.Son ideales cuando los datos pueden crecer, achicarse o modificarse.

Diccionarios: Se usan cuando necesitamos relacionar una clave con un valor.Son utiles cuando los valores tienen significado (por ejemplo:nombre,edad, nota).

Conjuntos (sets): Se usan cuando queremos guardar valores sin repetir y sin importar el orden. Son utiles para comparar grupos y eliminan duplicados.

```python
lista = [1, 2, 3]
```

```python
tupla = tuple(lista)
tupla
```
